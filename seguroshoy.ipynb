```python

import pandas as pd
data = pd.read_excel("seguroporhoy.xlsx", sheet_name="infoauto") #upload the file
filtro=pd.read_excel("Filtrados_info.xlsx")
data=data[~data['Código'].isin(filtro['Code'])]
```

```python
data.describe() #head the data to see if was propely upload
```

```python
def models(x):
    x = x.split(' ')[0]  #create a function to extract from the models data the first word separeted by a space
    return x
```

```python
def models_specification(x):
    x = " ".join(x.split()[1:-1])
    if x=="":
        x="-"
    return(x)
```

```python
data_agrupada=data
data_agrupada['mod_abreviados']=data_agrupada['Modelo'].apply(models) #generate the pandas column "mod abreviados" with th information of the models
data_agrupada['Modelo']=data_agrupada['Modelo'].apply(models_specification)
```

```python
data_agrupada=data.groupby(['Marca','Modelo','mod_abreviados']).sum().reset_index() #here group the data to obtain the sum of the values of the year with the objetive of filter bellow
data_agrupada=data_agrupada.drop('Código', 1) 
```

```python
data_agrupada.head(10000) #head the data
```

```python
tabla=data_agrupada.melt(id_vars=["Marca", "Modelo","mod_abreviados"], 
        value_vars=[2020,2019,2018,2017,2016,2015,2014,2013,2012,2011,2010,2009,2008,2007,2006,2005,2004,2003,2002,2001,2000,1999,1998,1997,1996,1995,1994,1993,1992,1991]
                        )   #in this line the data is transformed to put the yers from columns to rows
```

```python
validos=tabla[tabla['value']>0]  #filter the data that are not zero
```

```python
validos.head()
```

```python
validos_agrupados=validos.groupby(['Marca','mod_abreviados','variable']).Modelo.apply(lambda x: ','.join(map(str, x))).reset_index()
validos_agrupados.head() #group the models, separate the data by,','
```

```python
validos_agrupados=validos_agrupados.groupby(['Marca','Modelo','mod_abreviados']).variable.apply(lambda x: ','.join(map(str, x))).reset_index()
validos_agrupados.head() #group the years, separate the data by ','
```

```python
validos_agrupados.columns = ['Marca', 'Modelos', 'mod_abreviados', 'Años_disponibles'] #change the names of the columns
```

```python
validos_agrupados.head()
```

```python
validos_agrupados=validos_agrupados.groupby(['Marca'], as_index = False).agg({'Modelos': (lambda x: ','.join(map(str, x))), 'mod_abreviados': (lambda x: ','.join(map(str, x))),'Años_disponibles': (lambda x: ','.join(map(str, x)))})
validos_agrupados.head() #finally group only the Marca tag, concatenated the Marca, Modelo and mod_abreviado separated by ','
```

```python
def duplicates(x):
    x = ','.join(set(x.split(',')))
    return x    #create a funtion to delete the duplicates in the lists of the columns
```

```python
validos_agrupados['Años_disponibles']=validos_agrupados['Años_disponibles'].apply(duplicates)
validos_agrupados['mod_abreviados']=validos_agrupados['mod_abreviados'].apply(duplicates)
validos_agrupados['Modelos']=validos_agrupados['Modelos'].apply(duplicates)
validos_agrupados['Suma']=0
validos_agrupados.head()    #apply the duplicates funtion on the columns "Años disponible", "mod_Abreviados", "Modelos"
```

```python
matriz_propiedad_excel = pd.DataFrame()   #definition of the final propiety table
matriz_propiedad_excel['Código Info']=0
matriz_propiedad_excel['Type']='variable'
matriz_propiedad_excel['SKU']=0
matriz_propiedad_excel['Name']=validos_agrupados['Marca']
matriz_propiedad_excel['Categories']='30 garage,7 garage'  
matriz_propiedad_excel['Parent']=0
matriz_propiedad_excel['Attribute 1 name']='anio'
matriz_propiedad_excel['Attribute 1 value(s)']=validos_agrupados['Años_disponibles']
matriz_propiedad_excel['Attribute 2 name']='Modelo'
matriz_propiedad_excel['Attribute 2 value(s)']=validos_agrupados['mod_abreviados']
matriz_propiedad_excel['Attribute 3 name']='Versión'
matriz_propiedad_excel['Attribute 3 value(s)']=validos_agrupados['Modelos']
matriz_propiedad_excel['Attribute 4 name']='Cobertura'
matriz_propiedad_excel['Attribute 4 value(s)']='7 días Robo e Incendio en Garage,30 días Robo e Incendio en Garage'
matriz_propiedad_excel['Regular price']=0
matriz_propiedad_excel['Sale price']=0
matriz_propiedad_excel['In stock?']=1
matriz_propiedad_excel['Published']=1
matriz_propiedad_excel['Visibility in catalog']='visible'
matriz_propiedad_excel['Suma']=0
matriz_propiedad_excel.head()
```
